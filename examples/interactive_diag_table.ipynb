# MOM6 Diagnostic Table Generator

Interactive tool for creating MOM6 `diag_table` files from `available_diags` output.

## Quick Start Guide

Follow these steps to create your custom `diag_table`:

1. **Load Data** - Run the cells below to parse your `available_diags` file
2. **Set Case Name** - Enter your CESM case name in the interactive interface
3. **Add Files** - Create output files (daily, monthly, static, etc.)
4. **Select Diagnostics** - Use checkboxes to add diagnostics to each file
5. **Preview** - Review your configuration before exporting
6. **Export** - Save your `diag_table` file

## Key Features

- Interactive UI with search and filtering
- Diagnostics organized by type
- Quick search by name or description
- Configure multiple output frequencies
- Live preview of diag_table
- 2D/3D filtering
